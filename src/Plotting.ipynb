# Plotting individual dots

```python
plt.plot(3,4, 'r^') # x coordinate and y coordinate and marker (with color prefix if reqd)
plt.plot(2,4.2,'bo')
```

Above are in same axes (same graph). If we want in seperate graphs, we do __plt.show()__. It indicates that any plots above this line will appear in same graph

```python
plt.plot(3,4, 'r^')
plt.show()

# fresh figure for subsequent plotting
plt.plot(1,3,'g^')
plt.plot(2,4.2,'bo')
plt.show()
```

Adding a 'label' parameter to add legend to the graph. To show it in the graph, put __plt.legend()__. It will show in order of linear appearing in the code

```python
plt.plot(3,4, 'r^', label='red triangle')
plt.plot(1,3,'g^', label='green triangle') # label just adds info. Won't be displayed yet
plt.legend() # this will display label information
plt.show()
```

Another way to add legend is to provide label info as list to the legend function. Be mindful of the order.

__list in legend() will override label info in the plot()___

```python
plt.plot(3,4, 'r^', label='red triangle')
plt.plot(1,3,'g^', label='green triangle') # label just adds info. Won't be displayed yet
plt.legend(['green triangle', 'red triangle']) # this will display label information
plt.show()
```

# Plotting lines

To draw a line, mention the start and end points of the line

```python
x = [0,1]
y = [0,2]

plt.plot(x, y, 'm') # m - magenta
plt.show()
```

```python
# adding more points - x1 y1, x2 y2, x3 y3 ...
x = [0,1,3]
y = [0,2,-1]

plt.plot(x, y, 'ms--') # m - magenta, s - shape (square here), -- - line pattern
plt.show()
```

# Subplots

Indexing  starts at 1

```python
plt.subplot(1,2,1) # 1 row, 2 cols, 1st index of the plot
plt.plot(np.random.randn(10))

plt.subplot(1,2,2) # 1 row, 2 cols, 2nd index
plt.plot(np.random.randn(10))

plt.show()
```

```python
plt.subplot(2,2,1) # 2 row, 2 cols, 1st index of the plot
plt.plot(np.random.randn(10))

plt.subplot(2,2,4) # 2 rows, 2 cols, 4th index
plt.plot(np.random.randn(10))

plt.show()
```

Instead of using subplots, we use a bit more complicated yet better method:

```python
fig, ax = plt.subplots(1,2, figsize=(15,3)) # 1 row and 2 cols
# figure size is (width, height) in inches usually

#returns figure and axes. Indexing starts at 0

x = np.arange(2, 20)

ax[0].plot(x, np.log(x),'k')
ax[0].plot(x, np.sqrt(x), 'g')


ax[1].plot(x,2*np.sqrt(x),'r')
plt.show()
```

```python
fig, ax = plt.subplots(2,2)

# since we have 4 plots, we have to mention as 2D array
ax[0,0].plot(np.random.randn(4,4))
ax[0,1].plot(np.random.randn(4,4))
ax[1,0].plot(np.random.randn(4,4))
ax[1,1].plot(np.random.randn(4,4))

plt.tight_layout() # makes sure there is no overlapping among axes
plt.show()
```

# Making graphs look nicer

```python
x = np.linspace(-3,3,101)

plt.plot(x,x,label='y=x')
plt.plot(x,x**2,label='y=x^2')
plt.plot(x,2*x,label='y=2x')
plt.plot(x,x**3,label='y=x^3')
plt.legend()

# adding labels to axes
plt.xlabel('X')
plt.ylabel('F(x)')
plt.title('Functions')

# limiting axes range
plt.xlim([-3,3])
plt.ylim([-3,3])

# plt.axis('square') # sets length of x and y axis as same
plt.gca().set_aspect('auto') # get current aspect. 
# parameter of set_aspect - ratio of distance along each unit in X axis to distance along each unit in y-axis
plt.show()
```

```python
fig, ax = plt.subplots(1)

ax.plot(x,x,label='y=x')
ax.plot(x,x**2,label='y=x^2')
ax.plot(x,2*x,label='y=2x')
plt.plot(x,x**3,label='y=x^3')
ax.legend()

# adding labels to axes, using set_x, set_y, set_title
ax.set_xlabel('X')
ax.set_ylabel('F(x)')
ax.set_title('Functions')

# limiting axes range, using set methods
ax.set_xlim([-3,3])
ax.set_ylim([-3,3])

plt.show()
```

# Seaborn

Works on top of matplotlib. Often goes hand-in-hand with pandas

```python
import pandas as pd
import seaborn as sns
```

```python
n = 200
D = np.zeros((n,2))

D[:,0] = np.linspace(0,10,n) + np.random.randn(n) # adding noise
D[:,1] = D[:,0]**2 + np.random.randn(n)*10 # noiser than 1st column

sns.jointplot(x=D[:,0], y=D[:,1])
# plt.plot(D[:,0], D[:,1], 'o')
plt.show()
```

```python
df = pd.DataFrame(data=D, columns=['var1','var2'])
```

```python
df
```

```python
sns.jointplot(x=df.columns[0],y=df.columns[1],data=df)
```

```python
x = np.linspace(-1,1,n)
y1 = x**2
y2 = np.sin(3*x)
y3 = np.exp(-10*x**2)

sns.scatterplot(x=y1,y=y2,hue=y3, legend=False, palette='Spectral')
plt.show()
```

```python
sns.regplot(x=df.columns[0],y=df.columns[1], data=df,color='green')
plt.title(f'Regression of {df.columns[0]} on {df.columns[1]}')
plt.show()
```

# Images

```python
m, n = 3, 5

M = np.random.randint(10, size= (m,n))
M
```

__plt.imshow()__ creates image to represent numbers. Lightest box represents higher number, darkest represents lower

```python
plt.imshow(M)

plt.text(0,1,'hi') # x, y, and value to print on top of plot

plt.colorbar() # represents relationship b/w color and number
plt.show()
```

```python
plt.imshow(M)

for i in range(m):
    for j in range(n):
        plt.text(i,j,M[i,j])

plt.colorbar() # represents relationship b/w color and number
plt.show()
```

Above is incorrect because x-axis in image corresponds to columns of matrix, and y-axis in image corresponds to row in matrx. Hence, it should be swapped:

```python
plt.imshow(M)

for i in range(m):
    for j in range(n):
        plt.text(j,i,str(M[i,j]), # <- swap i and j
                 size=20, # increase font size
                 horizontalalignment='center', # aligning text
                 verticalalignment='center') 

plt.colorbar() # represents relationship b/w color and number
plt.show()
```
